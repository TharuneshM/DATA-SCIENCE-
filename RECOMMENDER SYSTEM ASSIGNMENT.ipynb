```python
user_book_df = data.pivot_table(columns='User.ID',
                                 index='Book.Title',
                                 values='Book.Rating')
```

```python
user_book_df
```

```python
user_book_df.fillna(0,inplace=True)
user_book_df
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
user_sim
```

```python
book_titles = user_book_df.index
```

```python
data.rename({"Book.Title":"Book_Title"},axis=1,inplace=True)
```

```python
# Define function provide top recommendations
def recommender(book_names):  
    # Initialize recommendations output length of number of books
    recommendations = np.zeros(user_sim.shape[0])
    # Loop through the books titles that the user has rated
    for book in book_names:
        # Sum the correlations
        recommendations = recommendations + user_sim[list(book_titles).index(book)]
    # Convert recommendations into dataframe
    recommendations_df = pd.DataFrame({
            'Title': book_titles,
            'Recommendation': recommendations})
    # Remove book titles that the user has already rated
    recommendations_df = recommendations_df[~(recommendations_df.Title.isin(book_names))]
    # Sort from most correlated to least correlated
    recommendations_df = recommendations_df.sort_values(by=['Recommendation'], ascending = False)
    return recommendations_df
```

```python
# Define user
user = 8

# Get list of books user has rated
user_books = data[data["User.ID"] == user].Book_Title.tolist()  
print('books rated by user\n')
print(user_books)
# Get list of recommendations
recommendations = recommender(user_books)

# Print out top 10 recommendations
print('\nRecommendations for user\n')
print(recommendations.Title.head(10))
```
