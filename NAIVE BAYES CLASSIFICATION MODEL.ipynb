```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.naive_bayes import MultinomialNB 
from sklearn.naive_bayes import GaussianNB 
```

```python
train_data=pd.read_csv("SalaryData_Train.csv")
```

```python
test_data=pd.read_csv("SalaryData_Test.csv")
```

```python
train_data.info()
```

# EDA on train data

```python
train_data.head(5)
```

```python
train_data.isnull().sum()
```

```python
train_data.drop("native",axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
train_data["Salary"]=label_encoder.fit_transform(train_data["Salary"])
```

```python
train_data1=pd.get_dummies(train_data,columns=["workclass"])
```

```python
train_data2=pd.get_dummies(train_data1,columns=["education"])
```

```python
train_data2.info()
```

```python
train_data3=pd.get_dummies(train_data2,columns=["maritalstatus","occupation","relationship","race","sex"])
```

```python
train_data3
```

```python
train_data3["salary"]=train_data3["Salary"]
train_data3.drop("Salary",axis=1,inplace=True)
```

```python
train_data3["salary"].value_counts()
```

# EDA on test data

```python
test_data.isnull().sum()
```

```python
test_data.drop("native",axis=1,inplace=True)
```

```python
test_data1=pd.get_dummies(test_data,columns=["workclass","education"])
```

```python
test_data2=pd.get_dummies(test_data1,columns=["maritalstatus","occupation","relationship","race","sex"])
```

```python
test_data2["Salary"]=label_encoder.fit_transform(test_data2["Salary"])
```

```python
test_data2
```

```python
test_data2["salary"]=test_data2["Salary"]
```

```python
test_data2.drop("Salary",axis=1,inplace=True)
```

```python
train_x=train_data3.iloc[:,0:-1]
```

```python
train_y=train_data3.iloc[:,-1]
```

```python
test_x=test_data2.iloc[:,0:-1]
```

```python
test_y=test_data2.iloc[:,-1]
```

# Multinomial naive bayes

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
MB=MultinomialNB()
```

```python
MB.fit(train_x,train_y)
```

```python
y_pred=MB.predict(test_x)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_y,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y,y_pred)
```

# Gaussian naive bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
GB=GaussianNB()
```

```python
GB.fit(train_x,train_y)
```

```python
y_pred1=GB.predict(test_x)
y_pred1
```

```python
print(classification_report(test_y,y_pred1))
```

```python
confusion_matrix(test_y,y_pred1)
```

```python
multinomial_accuracy=np.mean(test_y==y_pred)*100
multinomial_accuracy
```

```python
gaussian_accuracy=np.mean(test_y==y_pred1)*100
gaussian_accuracy
```

# GAUSSIAN NAIVE BAYES CLASSIFICATION PERFORMS BETTER THAN MULTINOMIAL NAIVE BAYES CLASSIFICATION
